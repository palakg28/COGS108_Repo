## Abstract

This research project aims to investigate the relationship between the age of a player and their estimated impact on team performance in the NBA and WNBA, contextualized by the league‚Äôs shifting generational dynamics in the early 2020s. With the iconic stars of the 2010‚Äôs (e.g. LeBron James, Stephen Curry, Kevin Durant) confronting a career decline as a result of age, this research project aimed to see how a player‚Äôs age would affect their dynamic and contribution to their team. Our team wanted to see how various players would capitalize on transitioning into this new era of basketball all in hopes of answering the question fans, reporters, and even the players themselves have been asking: who will become the new face of the league?

To investigate the relationship between player age and scoring impact, we collected and cleaned publicly available datasets from Basketball Reference and FiveThirtyEight, covering per-game player and team stats for both the NBA and WNBA. We then created a custom metric called scoring impact which is defined as the proportion of a player's points to their team‚Äôs total points in a given season. For NBA data, we estimated total player points using points-per-36 minutes multiplied by games played, while for WNBA players, we approximated points using usage rate(the percentage of a team‚Äôs possessions that a player is directly involved in) and minutes played. We then merged player data with team totals across seasons to compute each player‚Äôs scoring contribution percentage. To visualize trends, we used LOWESS smoothing and binned means to model how scoring impact varied with age. Finally, we conducted statistical analysis to determine peak ages and decline rates post-peak for each league, revealing clear age-related trends and notable differences between the NBA and WNBA trajectories.

Our analysis revealed that both NBA and WNBA players experience a peak influenced by age. However, we observed a difference in their peak ages. NBA players tend to reach their peak in their late 20s while WNBA players have a slightly later peak. This algins with our expectations and may be explained by factors such as a fewer number of games, less physical demands, and potentially a slower rate of physical decline among WNBA players.

# Research Question

Our research question seeks to understand how a player‚Äôs age influences their estimated impact to their team‚Äôs overall performance, specifically measured as the proportion of individual points to total team points, within the context of the NBA and WNBA. The aim is to explore whether aging correlates with changes in individual scoring impact, and if this relationship varies between male and female professional basketball players. This involves considering potential differences in career trajectories, physical decline, experience-driven performance, and league structures across genders. By comparing both leagues, the study also implicitly investigates broader questions about longevity, role evolution, and value attribution in professional sports.

# Hypothesis


We hypothesize that a basketball player‚Äôs impact will increase in their early career, peak at a certain age, and decrease as they approach retirement. Young players typically take time to develop teamwork and experience, which may result in lower impact. As they gain experience, they not only improve on their personal skills but also teamwork, leading to a peak. Later on, a decline to their physical abilities due to energy decline and injuries can cause their impact to drop.

We expect this pattern to be similar in both the NBA and WNBA, with the peak slightly later or lasting slightly longer for WNBA. This is possibly due to fewer games leading to less physical strain, a less aggressive playing style, and the possibility that women may experience a slower physical decline compared to men.

```python
import pandas as pd
import os

input_file = "data/NBA Teams/nba-data-historical.csv"
chunk_folder = "nba_chunks"
os.makedirs(chunk_folder, exist_ok=True)
chunk_size = 100_000

sample_df = pd.read_csv(input_file, nrows=5)
print("Column names preview:", list(sample_df.columns))

for i, chunk in enumerate(pd.read_csv(input_file, chunksize=chunk_size, low_memory=False)):
    chunk_path = os.path.join(chunk_folder, f"chunk_{i}.csv")
    chunk.to_csv(chunk_path, index=False)
    print(f"Saved {chunk_path}")
```

```python
chunk_files = sorted([f for f in os.listdir(chunk_folder) if f.endswith('.csv')])
nba_chunks = []

for file in chunk_files:
    chunk_path = os.path.join(chunk_folder, file)
    chunk_df = pd.read_csv(chunk_path, low_memory=False)
    nba_chunks.append(chunk_df)

nba_df = pd.concat(nba_chunks, ignore_index=True)

nba_df.columns = nba_df.columns.str.strip()

nba_df.dropna(subset=['name_common', 'team_id'], inplace=True)

percent_cols = [col for col in nba_df.columns if col.endswith('%')]
for col in percent_cols:
    if nba_df[col].dtype == object:
        nba_df[col] = nba_df[col].str.replace('%', '', regex=False).astype(float)

numeric_cols = [
    'age', 'G', 'Min', 'MP%', 'MPG', 'P/36', 'TS%', 'A/36', 'R/36',
    'SB/36', 'TO/36', 'Raptor O', 'Raptor D', 'Raptor+/-', 'Raptor WAR',
    'PIE%', 'AWS%', 'USG%', 'AST%', 'TOV%', 'ORB%', 'DRB%', 'TRB%',
    'STL%', 'BLK%', 'ORtg', 'DRtg', '2P%', '3P%', 'FT%', '3PAr', 'FTAr', 'Pace +/-'
]

nba_df[numeric_cols] = nba_df[numeric_cols].apply(pd.to_numeric, errors='coerce')

nba_df.to_csv('data/NBA Teams/nba_player_stats_cleaned.csv', index=False)

print(f"Total rows: {nba_df.shape[0]}")
print(f"Total columns: {nba_df.shape[1]}")
```

```python
#Visualizations
import matplotlib.pyplot as plt
import seaborn as sns

nba_viz = nba_df.dropna(subset=['name_common', 'P/36', 'G'])
nba_viz['estimated_total_points'] = nba_viz['P/36'] * nba_viz['G']

top_players = nba_viz.groupby('name_common')['estimated_total_points'].sum().sort_values(ascending=False).head(20)
top_df = nba_viz[nba_viz['name_common'].isin(top_players.index)]

plt.figure(figsize=(12, 6))
top_players.plot(kind='bar', color='steelblue')
plt.title("Top 20 NBA Players by Estimated Total Points (P/36 √ó Games)")
plt.ylabel("Estimated Total Points")
plt.xlabel("Player Name")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

team_points = nba_viz.groupby('team_id')['estimated_total_points'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
team_points.plot(kind='bar', color='indianred')
plt.title("Estimated Total Points Scored by Team (P/36 √ó Games)")
plt.xlabel("Team ID")
plt.ylabel("Estimated Total Points")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Analysis of Player Age Impact on Team Performance in NBA vs WNBA

We performed an analysis focusing on how player age influences their scoring impact relative to team performance, comparing NBA and WNBA players

### Data Preparation

```python
import pandas as pd
import numpy as np

nba_subset = nba_df[['name_common', 'team_id', 'year_id', 'age', 'MP%', 'P/36', 'TS%', 'USG%', 'Raptor WAR', 'G']].copy()
nba_subset.rename(columns={
    'name_common': 'Player',
    'team_id': 'Tm',
    'year_id': 'year_ID',
    'MP%': 'MP_pct',
    'P/36': 'Points_per_36',
    'USG%': 'USG_pct',
}, inplace=True)

numeric_cols = ['age', 'MP_pct', 'Points_per_36', 'TS%', 'USG_pct', 'Raptor WAR', 'G']
for col in numeric_cols:
    nba_subset[col] = pd.to_numeric(nba_subset[col], errors='coerce')

nba_subset.dropna(subset=['Points_per_36', 'G', 'age'], inplace=True)
nba_subset['totalPoints'] = nba_subset['Points_per_36'] * nba_subset['G']
nba_subset.rename(columns={'age': 'Age'}, inplace=True)


nba_season_points = nba_subset.groupby(['Player', 'year_ID', 'Age', 'Tm'])['totalPoints'].sum().reset_index()

nba_team_season_points = nba_season_points.groupby(['Tm', 'year_ID'])['totalPoints'].sum().reset_index()
nba_team_season_points.rename(columns={'totalPoints': 'teamTotalPoints'}, inplace=True)

nba_merged = pd.merge(nba_season_points, nba_team_season_points, on=['Tm', 'year_ID'], how='left')

nba_merged['points_pct'] = nba_merged['totalPoints'] / nba_merged['teamTotalPoints']

nba_significant = nba_merged[nba_merged['totalPoints'] > 100].copy()
nba_significant['League'] = 'NBA'

wnba_players = pd.read_csv('data/WNBA Teams/wnba_player_stats_cleaned.csv')

wnba_players['USG_pct'] = pd.to_numeric(wnba_players['USG_pct'], errors='coerce')
wnba_players['MP'] = pd.to_numeric(wnba_players['MP'], errors='coerce')

wnba_players['points'] = wnba_players['USG_pct'] * wnba_players['MP']

wnba_team_season_points = wnba_players.groupby(['Tm', 'year_ID'])['points'].sum().reset_index()
wnba_team_season_points.rename(columns={'points': 'teamTotalPoints'}, inplace=True)

wnba_merged = pd.merge(wnba_players, wnba_team_season_points, on=['Tm', 'year_ID'], how='left')
wnba_merged['points_pct'] = wnba_merged['points'] / wnba_merged['teamTotalPoints']

wnba_significant = wnba_merged[wnba_merged['MP'] > 200].copy()
wnba_significant['League'] = 'WNBA'

wnba_significant = wnba_significant.rename(columns={'year_ID': 'year_ID', 'points_pct': 'points_pct', 'League': 'League'})
wnba_significant['Age'] = pd.to_numeric(wnba_significant['Age'], errors='coerce')  # if 'Age' exists, else calculate if possible

comparison_df = pd.concat([
    nba_significant[['Age', 'points_pct', 'League']],
    wnba_significant[['Age', 'points_pct', 'League']]
], ignore_index=True)

comparison_df['Age_Bin'] = pd.cut(comparison_df['Age'], bins=range(15, 51, 5))
```

##### CODE FOR LOWESS REGRESSION PLOT

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

def plot_smooth_curve(data, title, color, xlim):
    data_sorted = data.sort_values(by='Age')
    lowess = sm.nonparametric.lowess
    smoothed = lowess(data_sorted['points_pct'], data_sorted['Age'], frac=0.3)  # Adjust frac for smoothing
    
    plt.figure(figsize=(14, 7))
    plt.scatter(data['Age'], data['points_pct'], alpha=0.1, color=color)
    plt.plot(smoothed[:, 0], smoothed[:, 1], color='black', linewidth=2)
    plt.title(title, fontsize=16)
    plt.xlabel('Player Age', fontsize=14)
    plt.ylabel('Percentage of Team Points', fontsize=14)
    plt.ylim(0, 0.3)
    plt.xlim(xlim)
    plt.grid(True)
    plt.show()
```

##### CODE FOR BINNED MEANS

```python
import pandas as pd
import numpy as np

def plot_binned_means(data, title, color, xlim):
    data['age_bin'] = pd.cut(data['Age'], bins=np.arange(18, 51, 2))  # 2-year bins
    
    binned_means = data.groupby('age_bin')['points_pct'].mean().reset_index()
    binned_means['age_mid'] = binned_means['age_bin'].apply(lambda x: x.mid)
    
    plt.figure(figsize=(14, 7))
    plt.plot(binned_means['age_mid'], binned_means['points_pct'], marker='o', color=color)
    plt.title(title, fontsize=16)
    plt.xlabel('Player Age (Binned)', fontsize=14)
    plt.ylabel('Avg. % of Team Points', fontsize=14)
    plt.ylim(0, 0.3)
    plt.xlim(xlim)
    plt.grid(True)
    plt.show()
```

### Additional Statistical Analysis

```python
import numpy as np
from scipy import stats

def get_peak_age(data, league_name):
    grouped = data.groupby('Age')['points_pct']
    valid_ages = grouped.filter(lambda x: len(x) >= 5).groupby(data['Age']).mean()
    
    if valid_ages.empty:
        print(f"{league_name}: Not enough valid data to estimate peak age.")
        return None
    
    peak_age = valid_ages.idxmax()
    print(f"{league_name} peak contribution age: {peak_age:.1f} years")
    return peak_age

def get_post_peak_decline(data, peak_age, league_name):
    if peak_age is None:
        print(f"{league_name}: Peak age not available for slope calculation.")
        return
    
    post_peak = data[data['Age'] > peak_age]
    if len(post_peak) < 5:
        print(f"{league_name}: Not enough post-peak data for decline rate estimation.")
        return

    slope, _, _, p_value, _ = stats.linregress(post_peak['Age'], post_peak['points_pct'])
    
    if np.isnan(slope):
        print(f"{league_name}: Regression could not compute slope (check data).")
        return
    
    print(f"{league_name} annual decline rate after peak: {-slope*100:.2f}% points per year (p = {p_value:.3f})")

nba_peak = get_peak_age(nba_significant, "NBA")
get_post_peak_decline(nba_significant, nba_peak, "NBA")

wnba_peak = get_peak_age(wnba_significant, "WNBA")
get_post_peak_decline(wnba_significant, wnba_peak, "WNBA")
```

This analysis identifies the age at which players in each league contribute the most to their team‚Äôs total points, based on average scoring percentage by age. It also calculates the linear decline in scoring contribution for players older than this peak age. For both leagues, we include ages only with at least five players. We apply a linear regression to post-peak ages to estimate the slope of decline in scoring contribution, along with the associated p-value.

These results show that WNBA players do peak at an older age than NBA players on average. WNBA declines at a higher rate, which isn‚Äôt statistically significant. This may be due to an outlier in the dataset or not enough data for older players.

# Discussion and Conclusion

Our analysis of NBA and WNBA player data reveals that age significantly influences a player‚Äôs scoring impact, measured as the proportion of individual points to team total points. Consistent with prior research, we find that NBA players generally reach their peak scoring impact in their late 20s before experiencing a gradual decline, a pattern attributed to the interplay of physical maturation, accumulated experience, and eventual physical decline. Notably, our custom metric, which directly links a player‚Äôs points to their team‚Äôs overall output, offers a more transparent view of individual contribution than composite statistics like PER or Win Shares, which can obscure direct scoring influence. In the WNBA, while a similar trend is observed, the peak appears to occur slightly later or to be more sustained, possibly due to factors such as shorter seasons, less physical strain, and a playing style that may allow for a longer prime. Our findings suggest that league context, gender, and even team dynamics play crucial roles in shaping a player‚Äôs scoring impact over time.

This suggests that aging in professional basketball is not the only correlative factor in performance but is instead shaped by league-specific factors such as game volume, physical demands, and player development pathways. These findings challenge conventional assumptions about athlete aging and performance, particularly when comparing across genders and leagues. For coaches, analysts, and team managers, understanding these nuanced differences can inform smarter roster decisions, training programs, and workload management strategies.
